```python
import pandas as pd
import matplotlib.pyplot as plt
datasiswa = pd.read_csv ("C:\\Users\\Althariq Fairuz\\Documents\\ITB\\Pengkom\\Praktikum pengkom\\Praktikum pengkom 5\\H05.csv")
datasiswa.head()
```

```python
# Problem 1
# Histogram distribusi nilai matematika
mat_df = datasiswa['nilai_mat']
ax = mat_df.plot(kind='hist',rwidth=0.5)
ax.set_xlabel('Nilai Matematika')
ax.set_ylabel('Jumlah Nilai')
ax.set_title ('Histogram Nilai Matematika')
plt.show()
```

```python
# Problem 2
# Diagram batang horizontal mahasiswa tiap fakultas
fakultas = datasiswa['fakultas'].value_counts()
ax = fakultas.plot(kind='barh')
ax.set_xlabel('Jumlah Mahasiswa')
ax.set_ylabel('Fakultas')
ax.set_title ('Bar horizontal sebaran mahasiswa tiap fakultas')
plt.show()
```

```python
# Problem 3
# Diagram pie banyaknya mahasiswa masing masing fakultas
fakultas = datasiswa['fakultas'].value_counts()
ax = fakultas.plot(kind='pie', autopct="%1.0f%%",figsize= [20,10])
ax.set_title ('Pie chart sebaran mahasiswa tiap fakultas')
ax.set_ylabel('')
plt.show()
```
